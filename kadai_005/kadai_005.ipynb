```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2,3,tight_layout=True)


y00 = [1,2,-5,2]
axes[0,0].plot(y00)



x01 = np.linspace(0,10,100)
y01 = 2 + 2 * np.sin(2 * x01)
axes[0,1].plot(x01,y01)


x02 = np.random .normal(15,5,2000)
axes[0,2].hist(x02,bins=30)


np.random.seed(3)
x10 = 4 + np.random.normal(0, 2, 60)
y10 = 4 + np.random.normal(0, 2, len(x10))
axes[1,0].scatter(x10, y10)


plt.show()
```
